#### Helper Functions

```python
def detect_outliers(column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_limit = Q1 - 1.5*IQR
    upper_limit = Q3 + 1.5*IQR
    outliers = (data[column] < lower_limit) | (data[column] > upper_limit)
    return outliers
```

```python
# Variance
numeric_cols = data.select_dtypes(include=['int64', 'float64'])
variance_values = numeric_cols.var()

# Skewness
skewness_values = numeric_cols.skew()

# Kurtosis
kurtosis_values = numeric_cols.kurtosis()

Stat_table = pd.DataFrame({'Column': variance_values.index,
                           'Variance': variance_values.values,
                            'Skewness': skewness_values.values,
                            'Kurtosis': kurtosis_values.values})

Stat_table
```

```python
#Covariance matrix
covar_matrix = numeric_cols.cov()
print("Covariance Matrix:\n", covar_matrix)
```

```python
#correlation using heatmap
corr_matrix = numeric_cols.corr()
print("Correlation Matrix:\n", corr_matrix)

plt.figure(figsize=(20, 20))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
def detect_outliers_z(column):
    z_scores = (data[column] - data[column].mean()) / data[column].std()
    return np.abs(z_scores) > 3
# Check for outliers
for column in ['lead time', 'average price ']:
    # Detect outliers
    outliers_z = detect_outliers_z(column)
    print(f"Number of outliers in {column} using Z-score:", np.sum(outliers_z))

     # Calculate the average (mean) of the column
    average_value = data[column].mean()

    # Handling outliers by replacing with the average
    data[column] = np.where(outliers_z, average_value, data[column])


# Check for remaining NaN values
print(data[['lead time', 'average price ']].isnull().sum())

# Visualizing the results
sns.boxplot(data['lead time'])
plt.title('Lead Time after Handling Outliers')
plt.show()

sns.boxplot(data['average price '])
plt.title('Average Price after Handling Outliers')
plt.show()


```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

X = data.drop('booking status', axis=1).dropna()
Y = data['booking status'].loc[X.index]

##applying LDA

lda = LDA(n_components=1)
X_lda = lda.fit_transform(X, Y)
X_lda
```

```python
#splitting data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import mean_absolute_error, mean_squared_error, accuracy_score, recall_score, f1_score, precision_score
#lda model
lda_model = LDA(n_components=1)
lda_model.fit(x_train,y_train)
y_pred = lda_model.predict(x_test)
# Evaluation
ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
plt.show()
LDA_acc = accuracy_score(y_test,y_pred)
precision = precision_score(y_test,y_pred)
recall = recall_score(y_test,y_pred)
f_score =  f1_score(y_test,y_pred)
mae = mean_absolute_error(y_test, y_pred)

evaluation_df = pd.DataFrame({
    'Accuracy': [LDA_acc],
    'Precision': [precision],
    'Recall/sensetivity': [recall],
    'F1 Score': [f_score],
    'Error Rate': [mae]
})
print("Evaluation data for LDA classifier " )
evaluation_df.T
```

```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.neighbors import KNeighborsClassifier
#KNN classifier

# Convert input data to NumPy arrays and ensure C-style contiguous
x_train = np.ascontiguousarray(x_train)
y_train = np.ascontiguousarray(y_train)
x_test = np.ascontiguousarray(x_test)

# Now you can proceed with fitting the model and making predictions
knn_model = KNeighborsClassifier(n_neighbors=3)
knn_model.fit(x_train, y_train)
y_pred = knn_model.predict(x_test)
# Evaluation
ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
plt.show()
KNN_acc = accuracy_score(y_test,y_pred)
precision = precision_score(y_test,y_pred)
recall = recall_score(y_test,y_pred)
f_score =  f1_score(y_test,y_pred)
mae = mean_absolute_error(y_test, y_pred)

evaluation_df = pd.DataFrame({
    'Accuracy': [KNN_acc],
    'Precision': [precision],
    'Recall/sensetivity': [recall],
    'F1 Score': [f_score],
    'Error Rate': [mae]
})
print("Evaluation data for KNN classifier " )
evaluation_df.T
```
