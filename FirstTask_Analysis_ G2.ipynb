# Hotel Data Analysis Report

## Overview

This report analyzes hotel booking data to uncover key insights and provide actionable recommendations. The focus is on understanding guest behaviors, cancellation patterns, and factors influencing booking retention. By analyzing guest profiles, booking trends, and various booking features, the report highlights potential areas of improvement to enhance customer retention and optimize services.

---

## Data Insights

### 1. Guest Profiles and Stay Patterns
- **Number of Adults**:
  - Range: 0 to 4 adults per booking.
  - Average: 1.84 adults.
  - 50% of bookings involve exactly 2 adults, most likely couples.
  
- **Number of Children**:

  - 75% of bookings have no children; most are for adults only.

---

### 7. Repeated Guests

- **Average**: 0.026, indicating that most guests are new.

---

### 8. Price Complaints
- **Range**: 0 to 13 complaints.
- **Average**: 0.023 complaints per booking.
- **Insight**: Most guests make no price complaints, but those who do tend to cancel at higher rates.

### 9. Non-Price Complaints

### 10. Meal Plans

### 11. Room Type

- **Car Parking Space**: Guests who didnâ€™t request parking are more likely to cancel.
- **Room Type**: Room Type 1 is the most popular but also has the highest cancellations.
- **Market Segment**: Online bookings have a higher cancellation risk compared to corporate and offline bookings.

---

## Summary of Cancellation Trends

Bookings are more likely to be canceled if they involve:
- A longer lead time.
- A higher average price per room.
- Fewer special requests.
- No car parking space.
- Meal Plan 1 or Room Type 1.
- Online booking methods.

---

## Recommendations

### 1. **Targeting Returning Guests**
- Implement loyalty programs or incentives to increase the number of repeat guests.
- Personalize offers and services to encourage returning guests, which would help improve retention.

### 2. **Flexible Cancellation Policies**
- Offer flexible cancellation policies, especially for high-priced or long-lead-time bookings.
- Implement small cancellation fees or offer incentives for guests to rebook rather than cancel.

### 3. **Promote Special Requests**
- Encourage guests to make special requests as part of their booking process. This can be promoted through marketing campaigns or added features in the booking process.

### 4. **Optimize Car Parking Space Offers**
- Promote parking options to reduce cancellation rates, particularly for guests who might not initially request parking.

### 5. **Reevaluate Meal Plan Options**
- Analyze the reasons for cancellations among guests choosing Meal Plan 1 and offer more flexible meal options.
- Provide special promotions for meal plans with lower cancellation rates.

### 6. **Repricing Strategy for Room Types**
- Adjust pricing strategies for Room Type 1, which has the highest cancellation rate.
- Promote Room Type 4 or other less-canceled room types with targeted marketing and price incentives.

### 7. **Corporate Booking Expansion**
- Corporate bookings show a lower cancellation rate. Expanding corporate booking options and offering packages for business travelers can increase retention.

### 8. **Reduce Online Booking Cancellations**
- Since online bookings have a higher cancellation risk, consider offering online-exclusive promotions or discounts to reduce cancellations. Additionally, introduce guarantees or flexible booking options to secure these reservations.

```python
# Normalize price per night
not_canceled["price"] = not_canceled["average_price_"] / (not_canceled["number_of_adults"] + not_canceled["number_of_children"])

# Aggregate room type columns into a single 'room_type' column
room_type_columns = [
    "room_Room_Type 1", "room_Room_Type 2", "room_Room_Type 3",
    "room_Room_Type 4", "room_Room_Type 5", "room_Room_Type 6",
    "room_Room_Type 7"
]

# Create a new DataFrame for room prices
room_prices = pd.melt(not_canceled, id_vars=["price"], value_vars=room_type_columns,
                      var_name="room_type", value_name="has_room_type")

# Filter only rows where the room type is present
room_prices = room_prices[room_prices["has_room_type"] == 1]

# Adjust the 'room_type' values for better readability
room_prices["room_type"] = room_prices["room_type"].str.replace("room_", "").str.replace("_", " ")

# Create the boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(x="room_type", y="price", data=room_prices, fliersize=0)
plt.title("Price of Room Types per Night and Person", fontsize=16)
plt.xlabel("Room Type", fontsize=16)
plt.ylabel("Price [EGP]", fontsize=16)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```
